```python
# Clean column names
df.columns = [col.strip().replace(' ', '_').replace('(', '').replace(')', '') for col in df.columns]
```

This line cleans column names ‚Äî spaces and brackets are removed, underscores are added. It helps make the code easier to write and read.

```python
df.head()
```

```python
# Convert date columns
df['Departure_DateTime'] = pd.to_datetime(df['Departure_Date_&_Time'])
df['Arrival_DateTime'] = pd.to_datetime(df['Arrival_Date_&_Time'])
```

 Converts string dates to actual datetime objects. This is essential to extract:

Time of day

Day of week

Month

Duration etc.

```python
df.tail()
```

```python

# Feature extraction
df['Departure_Hour'] = df['Departure_DateTime'].dt.hour
df['Departure_DayOfWeek'] = df['Departure_DateTime'].dt.day_name()
df['Departure_Month'] = df['Departure_DateTime'].dt.month_name()
```

From the Departure_DateTime, we extract:

Hour ‚Üí To analyze time-based pricing

Day of week ‚Üí Weekends vs weekdays

Month ‚Üí Seasonality

# Univariate Analysis
Analysis of one variable at a time to understand its basic characteristics like distribution, spread, central tendency (mean, median), and outliers.





```python
print(df.describe())
```

```python
## Histograms
#Purpose: Understand distribution of each variable individually.
num_cols = ['Total_Fare_BDT', 'Base_Fare_BDT', 'Tax_&_Surcharge_BDT', 'Duration_hrs', 'Days_Before_Departure']
for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

 Histograms of Total Fare, Base Fare, Tax & Surcharge
 
Skewed Right: Most fares are on the lower end, but there are some super expensive flights.

Shows the spread of flight prices.

Most fares are on the lower end.

Some flights are very expensive ‚Üí right-skewed distribution.

This indicates price outliers and variability in flight pricing.




```python
## Box plots
for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

Box Plots

Show median fare, IQR (Interquartile Range), and outliers clearly.

Flights with many stopovers or premium class tend to have higher variability in price.

Median fare (middle line)

Range of typical fares (the box)

Outliers (dots beyond whiskers)

Boxplots help us visually detect extreme values in pricing and price consistency.



# Bivariate Analysis
Analysis of two variables together to check their relationship or association.



```python
## Scatter plots
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='Days_Before_Departure', y='Total_Fare_BDT')
plt.title('Fare vs Days Before Departure')
plt.show()
```


‚óè Scatter Plot: Days_Before_Departure vs Total_Fare_BDT

Purpose: See if earlier bookings are cheaper.

What it shows: Weak negative trend ‚Äî booking earlier tends to cost less.

Insight: Not guaranteed savings; other factors (like season, airline) influence it.

```python
#box plot
plt.figure(figsize=(14, 6))
sns.boxplot(x='Airline', y='Total_Fare_BDT', data=df)
plt.xticks(rotation=45)
plt.title('Total Fare by Airline')
plt.xlabel('Airline')
plt.ylabel('Fare (BDT)')
plt.show()
```

 What it shows: Fare spread for each airline ‚Äî great to compare low-cost vs premium airlines.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Class', y='Total_Fare_BDT', data=df, palette='Set2')
plt.title('Total Fare by Travel Class')
plt.xlabel('Class')
plt.ylabel('Fare (BDT)')
plt.show()
```

 What it shows: Clear difference between Economy and Business class pricing.

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='Stopovers', y='Total_Fare_BDT', data=df, palette='Set3')
plt.title('Fare Distribution by Number of Stopovers')
plt.xlabel('Number of Stopovers')
plt.ylabel('Fare (BDT)')
plt.show()
```

 What it shows: More stopovers = cheaper flights (in general).

Boxplot: Stopovers vs Fare
Purpose: Compare fare across stopover counts.

What it shows: More stopovers = generally cheaper fares.

Insight: People may pay more for direct flights (time > money).


```python
## Cross-tabulations
cross_tab = pd.crosstab(df['Class'], df['Stopovers'])
print("\nCross Tabulation (Class vs Stopovers):")
print(cross_tab)
```


‚óè Cross-tabulation: Class vs Stopovers

Purpose: See how classes and stopover patterns relate.

What it shows: Economy = more stopovers. Business = fewer stopovers.

Insight: Premium classes prioritize comfort and direct routes.



# Multivariate Analysis

```python
sns.heatmap(df[num_cols].corr(), annot=True)
plt.title("Heatmap of Numerical Features")
```

 Heatmap of Correlation Matrix
 
Purpose: See how numeric variables relate to each other.

What it shows: Base Fare + Tax are highly correlated with Total Fare (logical).

Insight: Validates dataset integrity. Also, Days Before Departure has a slight negative correlation.

```python
## Pair plot
sns.pairplot(df[num_cols])
plt.show()
```

Pairplot of all numeric columns

Purpose: Visual overview of all variable pairs.

What it shows: Clusters, correlations, and distribution patterns.

Insight: Useful to spot outliers and groupings (e.g., short vs long flights).

```python
## Group analysis
grouped = df.groupby(['Airline', 'Class'])['Total_Fare_BDT'].mean().unstack()
print("\nAverage Fare by Airline and Class:")
print(grouped)
grouped.plot(kind='bar', figsize=(12, 6))
plt.title("Average Fare by Airline and Class")
plt.ylabel("Average Total Fare (BDT)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Grouped Bar Plot: Airline + Class vs Average Fare

Purpose: Compare how different airlines charge for economy vs business.

What it shows: Some airlines have a huge jump from economy to business. Others, not so much.

Insight: Shows pricing strategy and helps understand brand positioning of airlines.
